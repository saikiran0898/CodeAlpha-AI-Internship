```python
from langdetect import detect

texts = [
    "Hello everyone",
    "నమస్కారం",
    "नमस्ते",
    "Bonjour",
    "Hola"
]

languages = {
    "en": "English",
    "te": "Telugu",
    "hi": "Hindi",
    "fr": "French",
    "es": "Spanish"
}

for text in texts:
    code = detect(text)
    print(f"Text: {text}")
    print(f"Detected Language: {languages.get(code, code)}")
    print("-" * 30)
```
