```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer
```

```python
text = """
Artificial Intelligence is transforming industries around the world.
It is used in healthcare, education, banking, agriculture, and transportation.
AI helps automate repetitive tasks, improves decision-making, and increases efficiency.
Machine Learning and Deep Learning are important branches of Artificial Intelligence.
The future of AI is expected to create new opportunities and innovations across many sectors.
"""
```

```python
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
parser = PlaintextParser.from_string(text, Tokenizer("english"))
summarizer = LsaSummarizer()

print("Original Text:\n")
print(text)

print("\nSummary:\n")

for sentence in summarizer(parser.document, 2):
    print(sentence)
```
